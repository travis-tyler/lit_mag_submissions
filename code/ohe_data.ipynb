```python
import pandas as pd
import numpy as np
import ast
```

```python
df = pd.DataFrame(pd.read_csv('lit_mag_data.csv'))
df
```

```python
df['response_days'] = df.loc[df.response_time != 'No Data']['response_time'].str.split(' ').str[0]
df['response_days'] = df['response_days'].astype('float')
df = df.drop(columns=['response_time'])
df
```

```python
all_lengths = []
all_genres = []
for index, row in df.iterrows():
    [all_lengths.append(length) for length in ast.literal_eval(row['lengths']) if length not in all_lengths]
    [all_genres.append(genre) for genre in ast.literal_eval(row['genres']) if genre not in all_genres]
```

```python
row_lengths = []
row_genres = []
for index, row in df.iterrows():
    lenth_dict = {}
    genre_dict = {}
    for x in all_lengths:
        lenth_dict[f'length_{x}'] = 0
        if x in ast.literal_eval(row['lengths']):
            lenth_dict[f'length_{x}'] = 1
    row_lengths.append(lenth_dict)      
    for x in all_genres:
        genre_dict[f'genre_{x}'] = 0
        if x in ast.literal_eval(row['lengths']):
            genre_dict[f'genre_{x}'] = 1     
    row_genres.append(genre_dict)      

genre_df = pd.DataFrame(row_genres)
length_df = pd.DataFrame(row_lengths)
genre_df
```

```python
df = df.join(length_df)
df = df.join(genre_df)
df
```

```python
df.to_csv('ohe_lit_mag_data.csv')
```
