```python
import selenium, time, re, json
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.chrome.options import Options
import pandas as pd
import numpy as np

options = Options()
options.add_argument('--headless')
options.add_argument('--disable-gpu')
```

```python
url = 'https://thegrinder.diabolicalplots.com/Search/ByAlpha'

def browser():
    driver = webdriver.Chrome(ChromeDriverManager().install())
    driver.get(url)
    return driver
```

```python
def main(links):
    '''click through letter pages and copy tables''' 
    driver = browser()
    
    buttons = [driver.find_element_by_link_text(let) for let in links]
#     table_text = []
    table_html = []
        
    for button in buttons:
        time.sleep(3)
        driver.execute_script("arguments[0].click();", button)
#         tables.append(driver.find_element_by_tag_name("tbody").text)
        table_html.append(driver.find_element_by_tag_name("tbody").get_attribute('innerHTML'))
    
    all_tables = ''.join(table_html[1:])
    split_tables = all_tables.split('<tr')[1:]
    
    all_rows = []
    for row in split_tables[1:]:
        cells = row.split('<td role="gridcell')

        market_name = re.search('[0-9]">(.*)</a></td>', cells[1]).group(1)
        status = re.search('alt="(.*)" title=', cells[2]).group(1)
        genres = re.findall(r'alt="(.*?)"', cells[3])
        lengths = re.findall(r'alt="(.*?)"', cells[4])
        response_time = re.search('>(.*)</td></tr>', cells[5]).group(1)

        dic = {'market_name':market_name, 
               'status':status,
               'genres':genres, 
               'lengths':lengths,
               'response_time':response_time}

        all_rows.append(dic)

    return all_rows
```

```python
links = '#ABCDEFGHIJKLMNOPQRSTUVWXYZ'
data = main(links)
df = pd.DataFrame(data)
df.to_csv('../output/lit_mag_data.csv', index=False)
df
```
